```python
#Q1 Data Loading

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crypto_df = pd.read_csv('coinmarketcap_06122017.csv')
```

```python
crypto_df.head()
```

```python
#Q2 do count on id and market cap usd

crypto_df[['id','market_cap_usd']].count()
```

```python
# so from this we analyze that there are some coins which have zero values so we dont need them
#Q3
needed_df = crypto_df[crypto_df['market_cap_usd'] > 0]
```

```python
needed_df[['id','market_cap_usd']].count()
```

```python
# now we have valid data
```

```python
#Q4

top_ten = needed_df.nlargest(10,'market_cap_usd')
```

```python
top_ten
```

```python
# now we have the top 10 results so we can now plot them
```

```python
top_ten.plot(x='id',y='market_cap_usd',kind='bar')
```

```python
# so here the y axis is not clear so to make it clear we can convert these numbers in percent
```

```python
top_ten['market_top10_percentage'] = top_ten['market_cap_usd'] / top_ten['market_cap_usd'].sum() * 100
```

```python
top_ten
```

```python
#since now we have percentages we can easily plot them
```

```python
top_ten.plot(x='id',y='market_top10_percentage',kind='bar')
```

```python
# so now we have the difference. we can see bitcoin is miles ahead of all others
```

```python
#Q5 here to get a more better visualization we will convert the grph to log so we get much better readable results
top_ten.plot(x='id',y='market_cap_usd',kind='bar',log=True,color=['blue','red','blue','purple','purple','black','green','blue','teal','red'])
```
